```python
import requests 
from io import BytesIO 
from PIL import Image 

R = Image.open(BytesIO(requests.get('http://www.pythonchallenge.com/pc/def/oxygen.png').content)) 
(width, height) = R.size 
print('Width : %s\nHeight : %s' % (width, height)) 

Result = [] 
for i in range(0, width, 7): 
    Temp = chr(R.getpixel((i, 43))[0]) 
    print(Temp, end='') 
    Result.append(Temp) 
print('') 
Temp = ''.join(Result) 
Temp = (Temp[Temp.find('[')+1:Temp.find(']')]) 
for i in (Temp.split(', ')): 
    print(chr(int(i)), end='')

```
